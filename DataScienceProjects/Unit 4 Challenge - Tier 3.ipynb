```python
import matplotlib.pyplot as plt

# Filter the DataFrame for the specified year
df_2000 = df1[df1['Year'] >= 2000]

# Assuming your DataFrame is df_2000.describe()
top_means = df_2000.groupby('London_Borough')['Average_price'].mean().nlargest(6)
top_std = df_2000.groupby('London_Borough')['Average_price'].std().nlargest(6)
print(top_means)

# Plotting top standard deviations
top_std.plot(kind='bar', color='blue')
plt.xlabel('Borough')
plt.ylabel('Standard Deviation')
plt.title('Top Standard Deviations of Housing Prices (Past Two Decades)')
plt.show()

# Plotting top means
top_means.plot(kind='bar', color='red')
plt.xlabel('Borough')
plt.ylabel('Mean Price')
plt.title('Top Mean Housing Prices (Past Two Decades)')
plt.show()
```

```python
# Using the function 'groupby' will help you to calculate the mean for each year and for each Borough. 
## As you can see, the variables Borough and Year are now indices
dfg = df1.groupby(by=['London_Borough', 'Year']).mean()
dfg.sample(10)
```

```python
# Let's reset the index for our new DataFrame dfg, and call the head() method on it. 
dfg = dfg.reset_index()
dfg.head()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(london_borough, dataset):
    borough_data = dataset[dataset['London_Borough'] == london_borough]
    y1998 = float(borough_data['Average_price'][borough_data['Year'] == 1998])
    y2018 = float(borough_data['Average_price'][borough_data['Year'] == 2018])
    ratio = y2018 / y1998
    return ratio

# Test out the function
london_borough = 'Barking & Dagenham'
price_ratio = create_price_ratio('Barking & Dagenham', dfg)
print(f"Price Ratio for {london_borough}: {price_ratio}")

```

```python

london_boroughs = dfg['London_Borough'].unique()

# Create a DataFrame with all London Boroughs
boroughs_df = pd.DataFrame({'London_Borough': london_boroughs})

# Create a new column 'Price_Ratio' using the function
boroughs_df['Price_Ratio'] = boroughs_df['London_Borough'].apply(lambda borough: create_price_ratio(borough, dfg))

df_ratios = boroughs_df.sort_values(by='Price_Ratio', ascending=False)
```

```python
# Let's just rename the 'index' column as 'London_Borough', and the '0' column to '2018'.
df_ratios.rename(columns={'London_Borough':'Borough', 'Price_Ratio':'2018'}, inplace=True)
df_ratios.head()
```

```python
top15 = df_ratios.head(15)
 
ax = top15[['Borough','2018']].plot(kind='bar')

ax.set_xticklabels(top15.Borough)
```
