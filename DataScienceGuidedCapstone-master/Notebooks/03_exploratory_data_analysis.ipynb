One way to disentangle this interconnected web of relationships is via [principle components analysis](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html#sklearn.decomposition.PCA) (PCA). This technique will find linear combinations of the original features that are uncorrelated with one another and order them by the amount of variance they explain. You can use these derived features to visualize the data in a lower dimension (e.g. 2 down from 7) and know how much variance the representation explains. You can also explore how the original features contribute to these derived features.

The basic steps in this process are:

1. scale the data (important here because our features are heterogenous)
2. fit the PCA transformation (learn the transformation from the data)
3. apply the transformation to the data to create the derived features
4. (optionally) use the derived features to look for patterns in the data and explore the coefficients

#### 3.5.3.2 Calculate the PCA transformation<a id='3.5.3.2_Calculate_the_PCA_transformation'></a>

**Principal Component Analysis (PCA)** is a statistical technique used to simplify and analyze complex data by transforming it into a set of new variables called principal components. The main goal of PCA is to capture as much of the variability in the original data with as few components as possible.


* **Dimensionality Reduction:** Imagine you have a lot of information about ski resorts, like the number of lifts, area for night skiing, ticket prices, etc. PCA allows us to condense all this information into a smaller set of key factors (principal components) that still capture the essential features of the data. It's like summarizing a big book with a few key points.

* **Uncorrelated Features:** The principal components are designed to be uncorrelated, meaning they are independent of each other. This helps us avoid redundancy in the data. Each principal component represents a different aspect of the information.

* **Variance Explanation:** PCA tells us how much of the total variability in the original data is explained by each principal component. We can see which components contribute the most to understanding the differences between ski resorts.

* **Pattern Recognition:** When we plot the data using the principal components, we might observe patterns or clusters. For example, certain states or groups of resorts might behave similarly in terms of the features we're analyzing.

In the context of ski resorts, PCA could reveal which features are most important in distinguishing one resort from another, or which states have similar characteristics in terms of skiing amenities. It's a powerful tool for simplifying complex datasets and gaining insights into the underlying structure of the information.

Fit the PCA transformation using the scaled data.

```python
state_pca = PCA().fit(state_summary_scale)
```

This resulting state_summary_pca_result would be a new DataFrame with the same number of rows as state_summary_scale but potentially fewer columns, where each column represents a principal component.

PCA is often used for dimensionality reduction, visualization, or to identify patterns in the data by capturing the most significant variations in the original features.

Plot the cumulative variance ratio with number of components.

```python
#Code task 6#
#Call the `cumsum()` method on the 'explained_variance_ratio_' attribute of `state_pca` and
#create a line plot to visualize the cumulative explained variance ratio with number of components
#Set the xlabel to 'Component #', the ylabel to 'Cumulative ratio variance', and the
#title to 'Cumulative variance ratio explained by PCA components for state/resort summary statistics'
#Hint: remember the handy ';' at the end of the last plot call to suppress that untidy output
# Creating a subplot with a specified figure size
plt.subplots(figsize=(10, 6))

# Plotting the cumulative sum of explained variance ratios obtained from PCA
# state_pca.explained_variance_ratio_ contains the variance explained by each principal component
# cumsum() calculates the cumulative sum of these variances, providing insight into the total variance explained as we consider more components
plt.plot(state_pca.explained_variance_ratio_.cumsum())

# Adding labels to the plot
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')

# Display the plot
plt.show()
```


The "Cumulative Explained Variance by Principal Components" is typically expressed as a percentage. It represents the cumulative proportion of the total variance in the data that is explained by including each additional principal component. Each principal component contributes a certain percentage of the overall variance.

For example, if the cumulative explained variance after considering the first principal component is 30%, it means that this single component accounts for 30% of the total variability in the data. If you add a second principal component, and the cumulative explained variance becomes 50%, it means that together, the first two components explain 50% of the variance.

In percentage terms, the cumulative explained variance is often presented in a plot or a table to help you decide how many principal components to retain. A common threshold is to retain enough components to capture a high percentage of the total variance, such as 95% or more. This ensures that the retained components collectively preserve the majority of the information in the original data.

For instance, if you see that the cumulative explained variance reaches 95% after considering the first 5 principal components, it suggests that these 5 components are sufficient to capture most of the variability in the dataset. The choice of the percentage threshold can vary based on the specific goals of the analysis and the desired level of dimensionality reduction.

In this code, state_pca.explained_variance_ratio_ is an array containing the proportion of the variance explained by each principal component. 


cumsum() is then used to obtain the cumulative sum of these explained variances. The resulting plot shows how much of the total variance in the data is explained as we consider an increasing number of principal components. 


IThe first two components seem to account for over 75% of the variance, and the first four for over 95%.

Apply the transformation to the data to obtain the derived features.

```python
#Code task 7#
#Call `state_pca`'s `transform()` method, passing in `state_summary_scale` as its argument
state_pca_x = state_pca.transform(state_summary_scale)
```

By transforming the data in this way, each row in state_summary_scale is represented by its corresponding values along the principal components. The resulting state_pca_x variable contains this transformed data.

This transformation is often done after fitting a PCA model when you want to represent your original data in a lower-dimensional space defined by the principal components. 

```python
state_pca_x.shape
```

Plot the first two derived features (the first two principle components) and label each point with the name of the state.

Take a moment to familiarize yourself with the code below. It will extract the first and second columns from the transformed data (`state_pca_x`) as x and y coordinates for plotting. Recall the state labels you saved (for this purpose) for subsequent calls to `plt.annotate`. Grab the second (index 1) value of the cumulative variance ratio to include in your descriptive title; this helpfully highlights the percentage variance explained
by the two PCA components you're visualizing. Then create an appropriately sized and well-labelled scatterplot
to convey all of this information.

```python
# Extracting the first principal component from the transformed data
x = state_pca_x[:, 0]

# Extracting the second principal component from the transformed data
y = state_pca_x[:, 1]

# Extracting the state labels from the original data
state = state_summary_index

# Calculating the percentage of variance explained by the second principal component
pc_var = 100 * state_pca.explained_variance_ratio_.cumsum()[1]

# Creating a subplot with a specified figure size
plt.subplots(figsize=(10, 8))

# Creating a scatter plot using the first and second principal components
plt.scatter(x=x, y=y)

# Adding labels to the plot
plt.xlabel('First component')
plt.ylabel('Second component')

# Adding a title to the plot indicating the percentage of variance explained by the second component
plt.title(f'Ski states summary PCA, {pc_var:.1f}% variance explained')

# Annotating each point on the scatter plot with the corresponding state label
for s, x, y in zip(state, x, y):
    plt.annotate(s, (x, y))

# Displaying the plot
plt.show()
```

```python
# Extracting the first and second principal components from the DataFrame
x = pca_df['PC1']
y = pca_df['PC2']

# Extracting the 'AdultWeekend' ticket prices, quartiles, and state labels from the DataFrame
price = pca_df['AdultWeekend']
quartiles = pca_df['Quartile']
state = pca_df.index

# Calculating the percentage of variance explained by the second principal component
pc_var = 100 * state_pca.explained_variance_ratio_.cumsum()[1]

# Creating a subplot with a specified figure size
fig, ax = plt.subplots(figsize=(10, 8))

# Looping through each quartile category and plotting the corresponding points
for q in quartiles.cat.categories:
    # Selecting points belonging to the current quartile
    im = quartiles == q
    # Scatter plot for the selected points with size proportional to 'AdultWeekend' prices
    ax.scatter(x=x[im], y=y[im], s=price[im], label=q)

# Adding labels to the x and y axes
ax.set_xlabel('First component')
ax.set_ylabel('Second component')

# Adding a legend to distinguish quartiles
plt.legend()

# Adding a title to the plot indicating the percentage of variance explained by the second component
ax.set_title(f'Ski states summary PCA, {pc_var:.1f}% variance explained')

# Annotating each point on the scatter plot with the corresponding state label
for s, x, y in zip(state, x, y):
    plt.annotate(s, (x, y))

# Displaying the plot
plt.show()
```

Now, you see the same distribution of states as before, but with additional information about the average price. There isn't an obvious pattern. In this representation of the ski summaries for each state, which accounts for some 77% of the variance, you simply do not seeing a pattern with price.

Having merged your state summary features into the ski resort data, add "state resort competition" features:

#### 3.5.5.3 Scatterplots of numeric features against ticket price<a id='3.5.5.3_Scatterplots_of_numeric_features_against_ticket_price'></a>

Correlations, particularly viewing them together as a heatmap, can be a great first pass at identifying patterns. But correlation can mask relationships between two variables. You'll now create a series of scatterplots to really dive into how ticket price varies with other numeric features.

```python
# Define a useful function to create scatterplots of ticket prices against desired columns
def scatterplots(columns, ncol=None, figsize=(15, 8)):
    # If the number of columns per row (ncol) is not specified, set it to the total number of columns
    if ncol is None:
        ncol = len(columns)
    
    # Calculate the number of rows needed based on the specified number of columns
    nrow = int(np.ceil(len(columns) / ncol))
    
    # Create a subplot grid with the specified figure size
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    
    # Adjust the space between subplots
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    
    # Iterate over the specified columns and create scatterplots against 'AdultWeekend' ticket prices
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        ax.scatter(x=col, y='AdultWeekend', data=ski_data, alpha=0.5)
        ax.set(xlabel=col, ylabel='Ticket price')
    
    # Hide any extra subplots if the number of specified columns is less than the total
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
#Code task 13#
#Use a list comprehension to build a list of features from the columns of `ski_data` that
#are _not_ any of 'Name', 'Region', 'state', or 'AdultWeekend'
features = [column for column in ski_data.columns if column not in ['Name', 'Region', 'state',  'AdultWeekend']]
```

```python
scatterplots(features, ncol=4, figsize=(15, 15))
```

In the scatterplots you see what some of the high correlations were clearly picking up on. There's a strong positive correlation with `vertical_drop`. `fastQuads` seems very useful. `Runs` and `total_chairs` appear quite similar and also useful. `resorts_per_100kcapita` shows something interesting that you don't see from just a headline correlation figure. When the value is low, there is quite a variability in ticket price, although it's capable of going quite high. Ticket price may drop a little before then climbing upwards as the number of resorts per capita increases. Ticket price could climb with the number of resorts serving a population because it indicates a popular area for skiing with plenty of demand. The lower ticket price when fewer resorts serve a population may similarly be because it's a less popular state for skiing. The high price for some resorts when resorts are rare (relative to the population size) may indicate areas where a small number of resorts can benefit from a monopoly effect. It's not a clear picture, although we have some interesting signs.

Finally, think of some further features that may be useful in that they relate to how easily a resort can transport people around. You have the numbers of various chairs, and the number of runs, but you don't have the ratio of chairs to runs. It seems logical that this ratio would inform you how easily, and so quickly, people could get to their next ski slope! Create these features now.

```python
ski_data['total_chairs_runs_ratio'] = ski_data.total_chairs / ski_data.Runs
ski_data['total_chairs_skiable_ratio'] = ski_data.total_chairs / ski_data.SkiableTerrain_ac
ski_data['fastQuads_runs_ratio'] = ski_data.fastQuads / ski_data.Runs
ski_data['fastQuads_skiable_ratio'] = ski_data.fastQuads / ski_data.SkiableTerrain_ac
```

```python
scatterplots(['total_chairs_runs_ratio', 'total_chairs_skiable_ratio', 
              'fastQuads_runs_ratio', 'fastQuads_skiable_ratio'], ncol=2)
```

At first these relationships are quite counterintuitive. It seems that the more chairs a resort has to move people around, relative to the number of runs, ticket price rapidly plummets and stays low. What we may be seeing here is an exclusive vs. mass market resort effect; if you don't have so many chairs, you can charge more for your tickets, although with fewer chairs you're inevitably going to be able to serve fewer visitors. Your price per visitor is high but your number of visitors may be low. Something very useful that's missing from the data is the number of visitors per year.

It also appears that having no fast quads may limit the ticket price, but if your resort covers a wide area then getting a small number of fast quads may be beneficial to ticket price.
